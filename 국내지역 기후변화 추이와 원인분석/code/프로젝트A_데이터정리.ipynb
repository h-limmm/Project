```python
df88_96 = pd.read_csv('프로젝트A/data/산업88_96.csv')    
df97_00 = pd.read_excel('프로젝트A/data/제조업_중분류별_사업체_및_종사자수_1997-2000.xlsx')
df01_06 = pd.read_csv('프로젝트A/data/산업01_06.csv')
df07 = pd.read_csv('프로젝트A/data/산업07.csv')
df08 = pd.read_csv('프로젝트A/data/산업08.csv')                          
df09_19 = pd.read_csv('프로젝트A/data/산업09_19.csv')     
                        
```

```python
df01_06
```

```python
def m_df(df,num):
    com_col_li = []
    p_col_li = []
    for i in df.columns:
        if '.' not in i:
            com_col_li.append(i)
        else:
            p_col_li.append(i)
    com_df = df.loc[[num],com_col_li].T[3:]
    com_df.rename(columns={num:'산업체수'},inplace=True)
    p_df = df.loc[[num],p_col_li].T
    p_df.rename(columns={num:'종사자수'},inplace=True)
    idx = []
    for i in p_df.index.values:
        idx.append(i.split('.')[0])
    p_df.set_index(keys=[idx],inplace=True)
    
    return com_df,p_df
```
